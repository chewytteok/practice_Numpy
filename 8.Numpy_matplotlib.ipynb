```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 그래프 데이터 생성 방법

```python
x = np.linspace(0, 10, 11)
y = x ** 2 + x + 2 + np.random.randn(11)

print(x)
print(y)
```

### 그래프 출력하는 방법
- plot함수(선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
    - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
    - 기본적으로 x, y에 해당하는 값이 필요하다.

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

### 그래프에 주석 추가 방법
- x, y 축 및 타이틀

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.plot(x, y)
```

- grid 추가 방법

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)
plt.plot(x, y)
```

- x,y 축 범위 지정 방법

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)
plt.xlim(0, 13)
plt.ylim(0, 30)
plt.plot(x, y)
```

### plot 함수 parameters
- 그래프의 형태에 대한 제어 가능
- [plot함수 document](https://matplotlib.org/3.3.1/contents.html#)

- 그래프 색상 변경

```python
plt.plot(x, y, 'r')
```

```python
plt.plot(x, y, 'k')
```

```python
#RGB도 사용가능
plt.plot(x, y, '#ff00ff')
```

- 그래프 선스타일 변경

```python
plt.plot(x, y, '-.')
```

```python
plt.plot(x, y, 'g^')
```

```python
plt.plot(x, y, 'm:')
```

- 그래프의 두께 변경하는 방법
    - linewidth  명시

```python
plt.plot(x, y, 'm:', linewidth = 3)
```

- keyword parameter 이용하여 모든 속성 설정
    - color
    - linestyle
    - marker
    - markerfacecolor
    - markersize 등등

```python
plt.plot(x, y, color = 'yellow',
         linestyle= '--', marker='^',
         markerfacecolor = 'blue',markersize = 4)
```

### subplot으로 여러 그래프 출력하는 방법
- subplot함수로 구획을 구별하여 각각의 subplot에 그래프를 출력할 수 있다.

```python
plt.subplot(2,2,1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(x, y, 'g')

plt.subplot(2, 2, 3)
plt.plot(y, x, 'k')

plt.subplot(2, 2, 4)
plt.plot(x, np.exp(x), 'b')
```

### hist 함수 사용 방법
- histogram 생성
- bins로 histogram bar 개수 설정

```python
data = np.random.randint(1, 100, size = 200)
print(data)
```

```python
plt.hist(data, bins= 13, alpha= 0.6)
# bins는 막대의 개수, 분포의 세밀함을 조절 가능하다.
#alpha는 투명하게 변경할수있는 값 조정
plt.xlabel('value')
plt.ylabel('count')
plt.grid(True)
```
